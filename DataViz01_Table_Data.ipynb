```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.github.com/pandas-dev/pandas/master/pandas/tests/io/data/csv/iris.csv')
df    #ความยาว ความกว้างกลีบเลี้ยงกลับดอก
```

```python
set(df['Name'])#อยากรุ้ว่าในคอลัมนั้นมีค่าอะไรบ้าง ใช้ set เพื่อให้โชว์ค่าที่ไม่ซ้ำกัน อยากรุ้ว่าดอกไอริสมีสามชื่อใช่มั้ย
```

```python
df['Name'].unique()
```

# Parallel Coordinates

```python
pd.plotting.parallel_coordinates(df, 'Name') #เนมคือให้จัดกลุ่มตามชื่อ df คือชื่อตาราง
#ไอริซ v ทั้งสองงมกลีบเลี้ยงและกลีบดอกยาวเหมือนๆกัน ส่วนไอลิซ s กลีบดอกจะเล็กกลีบเลี้ยงจะยาว
```

### select columns to show
 ตารางสามารถปรับแต่งได้

```python
pd.plotting.parallel_coordinates(df, 'Name',cols = ['SepalLength','PetalLength']) #colsคือใส่ตัวที่เราต้องการเปรียบเทียบ
```

```python
pd.plotting.parallel_coordinates(df, 'Name',cols = ['SepalLength','PetalLength','SepalWidth','PetalWidth'])
```

### change color
เปลี่ยนสีได้

https://matplotlib.org/stable/tutorials/colors/colors.html  ลิงค์นี้ใช้ดูสี

```python
pd.plotting.parallel_coordinates(df, 'Name', color=['c','m','y'])   #color ใช้เปลี่ยนสี ใส่ชื่อสี ใช้ตัวย่อของสีเช่น R G B
```

```python
pd.plotting.parallel_coordinates(df, 'Name', color=[[0.5,0,0.5],[0,0.2,0.8],[0.2,0.4,0.7]])  #ใช้ค่าของแสง RGB ผสมกัน ค่า 1 คือปล่อยเต็ม 0 คือไม่ปล่อยยเลย
```

```python
pd.plotting.parallel_coordinates(df, 'Name', color=['#A73B24', '#0BE9DC', '#88B46C'])  #html color
```

### Specify color by colormap
https://matplotlib.org/stable/tutorials/colors/colormaps.html
Screenshot 2022-09-21 142054.png

```python
pd.plotting.parallel_coordinates(df, 'Name', colormap='magma')
```

# Matplotlib

```python
from matplotlib import pyplot as plt  #เอาไว้พร็อตกราฟ
```

```python
import matplotlib  #ปรับแต่งขนาดกราฟ
```

## Heatmap matrix

```python
matplotlib.rcParams['figure.figsize']=[30,25] #กำหนดขนาดของ figure
plt.imshow(df.iloc[:,:-1],interpolation='nearest')   #
#40-50มีสีต่าง เรียงดอกไม้สามชนิด โชว์ค่าสีตามเลขในตาราง
```

```python
plt.imshow(df.iloc[:,:-1],interpolation='nearest',cmap = 'hot')
```

## Subplot

```python
matplotlib.rcParams['figure.figsize']=[20,20] #กำหนดขนาดของกราฟ
```

```python
plt.subplot(1,3,1)  # subplot(มีกี่แถว, มีกี่หลัก, เป็นตัวที่เท่าไหร่)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest')

plt.subplot(1,3,2)
plt.imshow(df.iloc[50:100,:-1],interpolation='nearest')  #ตัวที่ 50-100

plt.subplot(1,3,3)
plt.imshow(df.iloc[100:150,:-1],interpolation='nearest')   #ตัวที่ 100-150
#แท่งแรกอ่อนไปเข็ม แท่งสองอ่อนสลับเข็ม แท่งสามอ่อนสลับเ๘็มกว่าอันที่สอง
```

```python
plt.subplot(1,4,2)
plt.imshow(df.iloc[50:100,:-1],interpolation='nearest',alpha = 0.8)

plt.subplot(1,4,1)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest',cmap='hot')

plt.subplot(1,4,2)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest',cmap='hot',alpha = 0.6)

plt.subplot(1,4,3)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest',cmap='hot',alpha = 0.5)

plt.subplot(1,4,4)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest',cmap='hot',alpha = 0.8)

#alpha คือความโปร่งใสของตัวรูป alpha=1 คือเต็ม ไม่โปร่งใสเลย
```

```python
plt.subplot(1,2,1)
pd.plotting.parallel_coordinates(df, 'Name', color=[[0.5,0.5,0.5,0.5],[0,0,0.8,0.5],[0.2,0.4,0.7,0.5]])

plt.subplot(1,2,2)
pd.plotting.parallel_coordinates(df, 'Name', color=[[0.5,0.5,0.5],[0,0,0.8],[0.2,0.4,0.7]])    #(R,G,B,alpha)
```
